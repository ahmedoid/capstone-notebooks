### تمرين 4: اختبار باستخدام assert

```python
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive")
        self.balance += amount

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive")
        if amount > self.balance:
            raise ValueError("Insufficient funds")
        self.balance -= amount

# TODO: اكتب اختبارات assert للتحقق مما يلي:
# 1. الحساب الجديد رصيده 0
# 2. بعد إيداع 100، يصبح الرصيد 100
# 3. السحب بمبلغ أكبر من الرصيد يرفع ValueError
# 4. الإيداع بمبلغ سالب يرفع ValueError
```
