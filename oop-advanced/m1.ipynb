## مفهوم الوراثة وعلاقة is-a
تنشأ علاقة الوراثة عندما نقول "السيارة هي مركبة" (Car is a Vehicle). في بايثون، نعبر عن ذلك بجعل الصنف `Car` يرث من `Vehicle`. يرث الصنف المشتق كل السمات والطرق العامة من الصنف الأساس، ويمكنه إضافة سمات جديدة أو تجاوز الطرق لتغيير السلوك.

```python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
    
    def move(self):
        print(f"The {self.make} {self.model} moves.")
    
    def __str__(self):
        return f"{self.year} {self.make} {self.model}"
```
