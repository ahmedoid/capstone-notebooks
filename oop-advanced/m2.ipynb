في التمرين التالي، لديك الصنف `A` (الأب) ثم `B` و `C` يرث كل منهما من `A`، وأخيرًا `D` يرث من `B` و `C`. ستكمل تعريف `B` و `C` ليستدعيا `super().info()` قبل طباعة اسميهما. الهدف هو أن عند استدعاء `d.info()` يُطبع: "D -> B -> C -> A". ثم اطبع `D.__mro__`.

```python
# تمرين: الماسة و MRO
# أكمل تعريف info في B و C ليستدعي super() قبل طباعة اسمه

class A:
    def info(self):
        print(' -> A', end='')

class B(A):
    def info(self):
        # TODO: استدعِ super().info() أولاً ثم اطبع " -> B"
        pass

class C(A):
    def info(self):
        # TODO: استدعِ super().info() أولاً ثم اطبع " -> C"
        pass

class D(B, C):
    def info(self):
        print(' -> D', end='')
        super().info()

# اختبر
d = D()
d.info()  # يجب أن يطبع:  -> D -> B -> C -> A
print()  # سطر جديد
# TODO: اطبع ترتيب MRO للصنف D
```
