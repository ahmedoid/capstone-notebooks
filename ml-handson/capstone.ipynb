```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# إعدادات الرسم
sns.set_style('whitegrid')
%matplotlib inline
```

```python
# توليد بيانات تخليقية لعقارات سعودية
# الميزات: المساحة (متر مربع)، عدد الغرف، وعمر العقار (سنوات)
# السعر: 500 * area + 1000 * rooms - 100 * age + ضوضاء

np.random.seed(42)
n = 300

area = np.random.uniform(50, 500, n)          # متر مربع
rooms = np.random.randint(1, 7, n)            # عدد الغرف
age = np.random.uniform(0, 50, n)             # عمر العقار

price = 500 * area + 1000 * rooms - 100 * age + np.random.normal(0, 20000, n)

df = pd.DataFrame({
    'area': area,
    'rooms': rooms,
    'age': age,
    'price': price
})
print(df.head())
```
