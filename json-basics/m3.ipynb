## 1. الأخطاء الشائعة (Common Errors)

JSON صارم في تركيبه. حتى خطأ بسيط يتسبب في `JSONDecodeError`. تشمل الأخطاء الشائعة:
- فاصلة زائدة بعد آخر عنصر في كائن أو مصفوفة (trailing comma).
- عدم إحاطة المفاتيح النصية بعلامات اقتباس مزدوجة (unquoted keys).
- قوس إغلاق من نوع غير مناسب (مثلاً `]` بدلاً من `}`).
في الخلية التالية، لدينا عدة سلاسل JSON خاطئة. مهمتك: اكتب حلقة تحاول تحميل كل سلسلة، وعند حدوث خطأ، اطبع معلومات موقعه.

## 2. التعامل مع JSONDecodeError

عند بناء تطبيقات تتعامل مع مدخلات المستخدم، من المهم عرض رسائل خطأ مفهومة. في هذه الخلية، ستكمل دالة `safe_parse` التي تحاول تحليل سلسلة JSON وتعيد :
- إذا نجحت: القاموس الناتج والرسالة `None`.
- إذا فشلت: `None` والرسالة تحتوي على وصف الخطأ.

```python
def safe_parse(json_string):
    """
    تحاول تحليل سلسلة JSON وإرجاع (data, None) إذا نجحت،
    وإلا (None, error_message).
    """
    try:
        data = json.loads(json_string)
    except json.JSONDecodeError as e:
        # TODO: كون رسالة خطأ تحتوي على رقم السطر والعمود والـ msg
        # استخدم return None, error_msg
        pass
    else:
        return data, None

test_strings = [
    '{"valid": "json"}',
    '{invalid}',
    '["list", "with", "extra",]',
    '"مجرد نص"'
]
for ts in test_strings:
    result, err = safe_parse(ts)
    if err is None:
        print(f"نجاح: {result}")
    else:
        print(f"فشل: {err}")
```
