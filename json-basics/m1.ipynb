## 1. استيراد مكتبة json

قبل البدء، نستورد المكتبة:

```python
import json

# JSON string representing a simple object
json_string = '{"name": "فاطمة", "age": 27, "is_student": false}'

# Parse the JSON string
parsed = json.loads(json_string)
print(parsed)
print(type(parsed))
```
