## استيراد المكتبات

سنحتاج إلى مكتبة json المضمنة في بايثون.

```python
import json
```

## تعريف بيانات JSON متداخلة

سننشئ قاموساً (Dictionary) يمثل بيانات شركة تحتوي على موظفين مع عناوين ونصوص عربية.

```python
# تعريف البيانات المتداخلة
complex_data = {
    "company": {
        "name": "شركة الأمل للتكنولوجيا",
        "location": "الرياض",
        "employees": [
            {
                "id": 1,
                "name": "أحمد محمد",
                "age": 30,
                "department": "تطوير",
                "contact": {
                    "email": "ahmed@example.com",
                    "phone": "0123456789"
                },
                "address": {
                    "street": "شارع الملك فهد",
                    "city": "الرياض",
                    "country": "السعودية"
                }
            },
            {
                "id": 2,
                "name": "سارة خالد",
                "age": 28,
                "department": "موارد بشرية",
                "contact": {
                    "email": "sara@example.com",
                    "phone": "0987654321"
                },
                "address": {
                    "street": "شارع الهرم",
                    "city": "الجيزة",
                    "country": "مصر"
                }
            }
        ]
    }
}
# طباعة جزء من البيانات للتأكد
print("اسم الشركة:", complex_data["company"]["name"])
print("عدد الموظفين:", len(complex_data["company"]["employees"]))
```

## التمرين 1: الوصول إلى البيانات المتداخلة

في هذا التمرين، ستتدرب على الوصول إلى قيمة محددة داخل الهيكل المتداخل باستخدام مفاتيح متعددة.
المطلوب: قم بتعيين المتغير `city` ليحتوي على مدينة الموظف الثاني (الموظف ذو `id=2`).

```python
# TODO: قم بتعيين المدينة الصحيحة بدلاً من None
city = None
print("مدينة الموظف الثاني:", city)
```

## التمرين 2: تنسيق JSON باستخدام indent

سنستخدم الدالة `json.dumps` لتحويل القاموس إلى سلسلة JSON منسقة. يمكننا تحديد عدد المسافات البادئة باستخدام الوسيط `indent`.
المطلوب: قم بتجربة قيم مختلفة لـ `indent` (مثل 2، 4، 8) ولاحظ الفرق. في الكود التالي، استبدل `0` برقم مناسب للمسافات البادئة.

```python
# TODO: استبدل 0 بقيمة مثل 4 للحصول على تنسيق جيد
indent_value = 0
json_str = json.dumps(complex_data, ensure_ascii=False, indent=indent_value)
print(json_str)
```

## التمرين 3: الحفاظ على النصوص العربية باستخدام ensure_ascii

عند تحويل النصوص إلى JSON، يقوم بايثون افتراضياً بترميز الحروف غير اللاتينية (مثل العربية) إلى رموز الهروب (\uXXXX). لمنع ذلك والحفاظ على النصوص كما هي، نستخدم `ensure_ascii=False`.
المطلوب: قارن بين المخرجات عند ضبط `ensure_ascii=True` (الافتراضي) و `ensure_ascii=False`.

```python
# TODO: في السطر التالي، تأكد من تعيين ensure_ascii=False
json_ascii = json.dumps(complex_data, ensure_ascii=True, indent=2)
print("مع ensure_ascii=True (افتراضي):")
print(json_ascii)
print("\n" + "="*50 + "\n")
json_no_ascii = json.dumps(complex_data, ensure_ascii=True, indent=2)  # TODO: غيّر True إلى False
print("مع ensure_ascii=False (لحفظ العربية):")
print(json_no_ascii)
```

## تمرين المراجعة

في هذا التمرين الشامل، ستقوم بما يلي:
- تحويل القاموس `complex_data` إلى سلسلة JSON مع `ensure_ascii=False` و `indent=2`.
- إعادة تحميل السلسلة إلى كائن بايثون باستخدام `json.loads`.
- إضافة موظف جديد إلى القائمة.
- تحويل الكائن المعدل مرة أخرى إلى JSON منسق.
المطلوب: أكمل الكود التالي بإكمال الحقول الفارغة وتعيين القيم المناسبة.

```python
# تحويل البيانات إلى JSON
json_original = json.dumps(complex_data, ensure_ascii=False, indent=2)
# تحميل النص JSON إلى كائن
data = json.loads(json_original)
# TODO: أكمل بيانات الموظف الجديد أدناه
new_employee = {
    "id": 0,
    "name": "علي حسن",
    "age": 0,
    "department": "",
    "contact": {
        "email": "",
        "phone": ""
    },
    "address": {
        "street": "",
        "city": "",
        "country": ""
    }
}
# إضافة الموظف الجديد
data["company"]["employees"].append(new_employee)
# TODO: عيّن قيمة indent المناسبة (مثلاً 2) بدلاً من 0 في السطر التالي
json_updated = json.dumps(data, ensure_ascii=False, indent=0)
print(json_updated)
```
