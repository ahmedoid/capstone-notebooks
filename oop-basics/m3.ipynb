## 3. استخدام المزين @property
حول دالة `get_name()` إلى خاصية للقراءة فقط باستخدام `@property`. بحيث يمكن الوصول إلى الاسم كخاصية وليس كدالة.

```python
class Student:
    def __init__(self, name):
        self.__name = None
        self.set_name(name)

    def set_name(self, new_name):
        if not new_name:
            raise ValueError('الاسم لا يمكن أن يكون فارغًا')
        self.__name = new_name

    # TODO: أضف المزين property لتحويل get_name إلى خاصية name (للقراءة فقط)
    def get_name(self):
        return self.__name

# اختبر الكود
s2 = Student('نور')
# TODO: اطبع اسم الطالب باستخدام الخاصية s2.name (وليس دالة)
```

## 4. خاصية قابلة للكتابة مع @setter
أضف إمكانية تعديل الاسم عبر الخاصية نفسها مع التحقق، باستخدام `@name.setter`.

```python
class Student:
    def __init__(self, name):
        self.__name = None
        self.name = name  # استخدم الخاصية مباشرة

    @property
    def name(self):
        return self.__name

    # TODO: أضف دالة setter باسم name باستخدام @name.setter مع نفس تحقيق السابق

# اختبر الكود
s3 = Student('ليلى')
print(s3.name)
# TODO: حاول تغيير الاسم إلى 'مريم' باستخدام العملية = ثم اطبع القيمة الجديدة. جرب تعيين اسم فارغ.
```
