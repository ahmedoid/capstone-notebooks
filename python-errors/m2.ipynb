# البنية الكاملة: else و finally وتعدد فروع الالتقاط

### 1. التذكير بالأساسيات: try-except-else-finally
البنية الكاملة تتكون من:
* `try`: الكود الذي قد يرفع استثناء.
* `except`: معالجة الاستثناء.
* `else`: ينفذ فقط إذا لم يحدث أي استثناء في كتلة `try`.
* `finally`: ينفذ دائمًا سواء حدث استثناء أم لا.

في الخلية التالية، قم بإكمال كتلتي `else` و `finally`.

```python
a = 10
b = 2  # جرب تغيير القيمة إلى 0 لملاحظة الفرق

try:
    result = a / b
except ZeroDivisionError:
    print("لا يمكن القسمة على صفر")
# TODO: أكمل else لطباعة نتيجة القسمة عند النجاح
# TODO: أكمل finally لطباعة 'تم التنفيذ' دائمًا
```

### 2. استخدام else للتشغيل عند النجاح
في المثال التالي، دالة تتحقق من العمر وترفع ValueError إذا كان سالبًا. أضف كتلة `else` لطباعة رسالة نجاح.

```python
def check_age(age):
    if age < 0:
        raise ValueError("العمر لا يمكن أن يكون سالبًا")
    return age

age_input = 25
try:
    age = check_age(age_input)
except ValueError as e:
    print("خطأ:", e)
# TODO: أضف else لطباعة "العمر صالح:" وقيمته
```
