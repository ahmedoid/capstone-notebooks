## 1. المجموعات (Groups) والبدائل (Alternation)

تسمح لنا الأقواس `()` بتجميع أجزاء من النمط لاستخراجها لاحقًا، بينما نستخدم الرمز `|` للبدائل لتعريف خيارات متعددة.

```python
import re

text = "Phone: 123-456-7890"
# TODO: اكتب نمطاً يستخدم المجموعات لاستخراج رمز المنطقة (أول ثلاث أرقام) والرقم المتبقي
pattern = r"..."  # استبدل ... بالنمط الصحيح
match = re.search(pattern, text)
if match:
    print("Area code:", match.group(1), "Number:", match.group(2))
```

## 2. التعامل مع المحارف الخاصة (Escaping)

بعض الرموز مثل `.` و `+` و `*` تعتبر محارف خاصة في التعابير النمطية. لمطابقتها بشكل حرفي، نستخدم الشرطة المائلة للخلف `\\`.

```python
import re

filename = "file.txt"
# TODO: اكتب نمطاً يطابق الملفات التي تنتهي بـ .txt مع الحرص على تطابق النقطة الحرفية (استخدم \\.)
pattern = r"..."
match = re.search(pattern, filename)
if match:
    print("تم العثور على", match.group())
```

## 3. استخدام `re.findall` و `re.sub`

`re.findall` يعيد قائمة بكل التوافقات، بينما `re.sub` يستبدل التوافقات بنص آخر.

```python
import re

text = "Contact us at info@example.com or support@company.org. For personal, email admin@site.net."
# TODO: استخدم re.findall مع نمط بسيط للبريد الإلكتروني (مثل \\S+@\\S+) لاستخراج جميع العناوين
emails = re.findall(r"...", text)
print(emails)
```

```python
import re

text = "Call 123-456-7890 or 987-654-3210 for assistance."
# TODO: استخدم re.sub لاستبدال أي رقم هاتف (نمط \\d{3}-\\d{3}-\\d{4}) بكلمة '[PHONE]'
new_text = re.sub(r"...", "[PHONE]", text)
print(new_text)
```
