# الكميات والتثبيت في التعابير النمطية

في هذا المختبر سنتدرب على استخدام الكميات `*` و `+` و `?` و `{n,m}` والمراسي `^` و `$`.

```python
import re
import numpy as np
np.random.seed(42)

# توليد نصوص اختبار عشوائية
test_strings = ["a", "ab", "abb", "abbb", "b", "aa", "aab", "cab"]
print("Test strings for quantifier practice:", test_strings)
```

## الكمية `*` (النجمة)

الكمية `*` تعني 'صفر أو أكثر' من العنصر الذي يسبقها. مثلاً، النمط `ab*` يطابق `a` متبوعاً بصفر أو أكثر من `b`.

```python
# TODO: عرّف النمط pattern ليطابق a متبوعًا بصفر أو أكثر من b باستخدام *
pattern = r""  # اكتب النمط هنا

print("Pattern:", pattern)
for s in test_strings:
    if re.fullmatch(pattern, s):
        print(f"'{s}' matches")
    else:
        print(f"'{s}' does NOT match")
```

## الكمية `+` (علامة الجمع)

الكمية `+` تعني 'واحد أو أكثر'. النمط `ab+` يطابق `a` متبوعاً بواحد أو أكثر من `b`.

```python
# TODO: عرّف النمط pattern ليطابق a متبوعًا بواحد أو أكثر من b باستخدام +
pattern = r""  # اكتب النمط هنا

print("Pattern:", pattern)
for s in test_strings:
    if re.fullmatch(pattern, s):
        print(f"'{s}' matches")
    else:
        print(f"'{s}' does NOT match")
```

## الكمية `?` (علامة الاستفهام)

الكمية `?` تعني 'صفر أو واحد'، أي أن العنصر اختياري. النمط `ab?` يطابق `a` أو `ab`.

```python
# TODO: عرّف النمط pattern ليطابق a متبوعًا بصفر أو واحد من b باستخدام ?
pattern = r""  # اكتب النمط هنا

print("Pattern:", pattern)
for s in test_strings:
    if re.fullmatch(pattern, s):
        print(f"'{s}' matches")
    else:
        print(f"'{s}' does NOT match")
```

## الكميات المحددة `{n}` و `{n,m}`

- `{n}`: تطابق العنصر عدداً محدداً من المرات (n مرة بالضبط).
- `{n,m}`: تطابق بين n و m مرة.

مثال: `a{3}` يطابق `aaa`، و `a{2,4}` يطابق `aa` أو `aaa` أو `aaaa`.

تطبيق: اكتب نمطاً يطابق كلمة تبدأ بحرفين كبيرين أو صغيرين، ثم ثلاثة أرقام بالضبط، ثم حرف واحد (مثل `AB123c`).

```python
# توليد نصوص اختبار للكميات المحددة
test_quant = ["AB123c", "ab123d", "A123c", "AB12c", "AB1234d", "XY999z"]
print("Test strings for {n,m}:", test_quant)

# TODO: اكتب نمطًا يطابق نمط: حرفين، ثم 3 أرقام، ثم حرفًا
pattern = r""  # اكتب النمط هنا

print("Pattern:", pattern)
for s in test_quant:
    if re.fullmatch(pattern, s):
        print(f"'{s}' matches")
    else:
        print(f"'{s}' does NOT match")
```

## المراسي (Anchors) `^` و `$`

- `$` يطابق نهاية النص.

باستخدامهما، يمكننا التأكد من أن النمط يطابق النص كاملاً من البداية إلى النهاية.

تطبيق: اكتب نمطاً يتحقق من أن النص يبدأ بـ `Mr.` وينتهي بـ `Smith` (يوجد نص بينهما).

```python
test_anchor = ["Mr. Smith", "Mr. John Smith", "Mr. Smithy", "Not Mr. Smith", "Mr.Smith"]
print("Test strings for anchors:", test_anchor)

# TODO: اكتب نمطًا يبدأ بـ 'Mr.' وينتهي بـ 'Smith'
pattern = r""  # اكتب النمط هنا

print("Pattern:", pattern)
for s in test_anchor:
    if re.fullmatch(pattern, s):
        print(f"'{s}' matches")
    else:
        print(f"'{s}' does NOT match")
```

## تمرين: التحقق من رقم الجوال السعودي

صيغ أرقام الجوال السعودية:
- 05xxxxxxxx: يبدأ بـ 05 ثم 8 أرقام (المجموع 10 أرقام)
- +9665xxxxxxxx: يبدأ بـ +966 ثم 5 ثم 8 أرقام
- 009665xxxxxxxx: يبدأ بـ 00966 ثم 5 ثم 8 أرقام

المطلوب: كتابة نمط يتحقق من صحة رقم الجوال، ويقبل الصيغ الثلاث.

```python
# قائمة أرقام للاختبار
phone_numbers = [
    "0512345678",   # صحيح - محلي
    "+966512345678",# صحيح - دولي مع +
    "00966512345678",# صحيح - دولي 00966
    "051234567",    # خاطئ - ناقص رقم
    "05123456789",  # خاطئ - زائد رقم
    "+96651234567", # خاطئ - ناقص
    "0096651234567",# خاطئ - ناقص
    "+961512345678",# خاطئ - رمز دولة خطأ
    "051234567o",   # خاطئ - حرف
]
print("Test phone numbers:")
for num in phone_numbers:
    print(f"  {num}")

# TODO: اكتب النمط الصحيح للتحقق من رقم الجوال السعودي
pattern = r""  # اكتب النمط هنا

print("\nPattern:", pattern)
for num in phone_numbers:
    if re.fullmatch(pattern, num):
        print(f"'{num}' is VALID Saudi mobile number")
    else:
        print(f"'{num}' is INVALID")
```
