```python
# استيراد المكتبات
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
## توليد البيانات التخليقية

سنقوم بتوليد مجموعة بيانات تحتوي على 200 عقار. المتغيرات المستقلة هي:
- `area`: مساحة العقار بالمتر المربع (بين 50 و 500 م²)
- `rooms`: عدد الغرف (بين 1 و 10)
- `age`: عمر العقار بالسنوات (بين 0 و 50 سنة)

سيتم حساب سعر تقريبي غير حقيقي باستخدام علاقة خطية مع ضوضاء عشوائية، ومن ثم تصنيف العقار إلى:
- `1` (غالي) إذا كان السعر أعلى من 500,000 ريال سعودي
- `0` (رخيص) إذا كان السعر أقل أو يساوي 500,000 ريال

**تنبيه:** هذه البيانات تخليقية بالكامل ولا تعكس أسعار السوق الفعلية.
```

```python
# إعداد البيانات التخليقية
np.random.seed(42)
n_samples = 200

# المساحة: توزيع منتظم بين 50 و 500
area = np.random.uniform(50, 500, n_samples)
# عدد الغرف: عدد صحيح بين 1 و 10
rooms = np.random.randint(1, 11, n_samples)
# العمر: توزيع منتظم بين 0 و 50
age = np.random.uniform(0, 50, n_samples)

# حساب السعر التقريبي بمعادلة خطية + ضوضاء
price = area * 5000 + rooms * 10000 - age * 2000 + np.random.normal(0, 50000, n_samples)

# التصنيف: 1 إذا السعر > 500,000
label = (price > 500000).astype(int)

# إنشاء DataFrame
df = pd.DataFrame({
    'area': area,
    'rooms': rooms,
    'age': age,
    'price': price,
    'label': label
})

print("تم إنشاء البيانات التخليقية.")
```
