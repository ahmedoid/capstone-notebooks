```python
import pandas as pd
import numpy as np
np.random.seed(42)

# إنشاء بيانات العملاء (Customers)
customers = pd.DataFrame({
    'customer_id': range(1, 6),
    'customer_name': ['أحمد', 'فاطمة', 'علي', 'نورة', 'محمد'],
    'city': ['الرياض', 'جدة', 'الدمام', 'جدة', 'الرياض']
})

# إنشاء بيانات المنتجات (Products)
products = pd.DataFrame({
    'product_id': range(101, 106),
    'product_name': ['كتاب', 'قلم', 'دفتر', 'حقيبة', 'آلة حاسبة'],
    'price': [25, 5, 10, 50, 30]
})

# إنشاء بيانات الطلبات (Orders) بشكل عشوائي
order_ids = range(1, 11)
customer_ids = np.random.choice(customers['customer_id'], size=10, replace=True)
product_ids = np.random.choice(products['product_id'], size=10, replace=True)
quantities = np.random.randint(1, 5, size=10)
dates = pd.date_range(start='2025-01-01', periods=10, freq='D')
ordered_dates = np.random.choice(dates, size=10, replace=False)

orders = pd.DataFrame({
    'order_id': order_ids,
    'customer_id': customer_ids,
    'product_id': product_ids,
    'quantity': quantities,
    'order_date': ordered_dates
})

# ضبط تنسيق التاريخ
orders['order_date'] = pd.to_datetime(orders['order_date'])
```
