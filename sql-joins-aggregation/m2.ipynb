# المختبر: التجميع والدوال المجمّعة في SQL

في هذا المختبر، ستطبق مفاهيم التجميع والدوال المجمّعة في SQL باستخدام مكتبة sqlite3 في بايثون. ستتدرب على كتابة استعلامات COUNT, SUM, AVG, MIN, MAX, GROUP BY, و HAVING. البيانات جميعها مولّدة ذاتيًا داخل الدفتر.

```python
import sqlite3
import numpy as np

np.random.seed(42)
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()
```

```python
# إنشاء جدول المنتجات (products) وإدخال بيانات عشوائية
cursor.execute('''
    CREATE TABLE products (
        product_id INTEGER PRIMARY KEY,
        product_name TEXT,
        category TEXT,
        price REAL
    )
''')

categories = ['Electronics', 'Clothing', 'Books', 'Home', 'Sports']
num_products = 20
prices = np.round(np.random.uniform(10, 500, num_products), 2)
cat_choices = np.random.choice(categories, num_products)

for i in range(num_products):
    name = f'Product_{i+1}'
    cursor.execute("INSERT INTO products (product_id, product_name, category, price) VALUES (?, ?, ?, ?)",
                   (i+1, name, cat_choices[i], prices[i]))
conn.commit()
print('تم إنشاء جدول المنتجات وإدخال البيانات.')
```

```python
# عرض أول 5 صفوف من جدول المنتجات
cursor.execute('SELECT * FROM products LIMIT 5')
rows = cursor.fetchall()
for row in rows:
    print(row)
```
