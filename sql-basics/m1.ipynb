البيانات المستخدمة تحاكي متجرًا إلكترونيًا سعوديًا يحتوي على جداول: المنتجات (products)، العملاء (customers)، والطلبات (orders).

## إعداد قاعدة البيانات

الخلية التالية تقوم بإنشاء قاعدة بيانات في الذاكرة وإنشاء الجداول وإدخال بيانات تجريبية. قم بتشغيلها مرة واحدة.

```python
import sqlite3
import numpy as np

np.random.seed(42)

# إنشاء اتصال بقاعدة بيانات في الذاكرة
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# إنشاء الجداول
cursor.executescript('''
CREATE TABLE products (
    product_id INTEGER PRIMARY KEY,
    name TEXT,
    category TEXT,
    price REAL,
    stock INTEGER
);

CREATE TABLE customers (
    customer_id INTEGER PRIMARY KEY,
    name TEXT,
    city TEXT,
    email TEXT
);

CREATE TABLE orders (
    order_id INTEGER PRIMARY KEY,
    customer_id INTEGER,
    order_date TEXT,
    total_amount REAL,
    FOREIGN KEY(customer_id) REFERENCES customers(customer_id)
);
''')

# إدخال بيانات المنتجات
product_names = ['هاتف ذكي', 'سماعات', 'شاحن', 'كيبورد', 'فأرة', 'حامل هاتف', 'كابل USB', 'باور بانك']
categories = ['إلكترونيات', 'إلكترونيات', 'إلكترونيات', 'كمبيوتر', 'كمبيوتر', 'إلكترونيات', 'إلكترونيات', 'إلكترونيات']
for i in range(len(product_names)):
    price = round(np.random.uniform(20, 500), 2)
    stock = np.random.randint(5, 50)
    cursor.execute("INSERT INTO products VALUES (?,?,?,?,?)", (i+1, product_names[i], categories[i], price, stock))

# إدخال بيانات العملاء
customer_names = ['أحمد', 'فاطمة', 'خالد', 'نورة', 'محمد']
cities = ['الرياض', 'جدة', 'الدمام', 'مكة', 'المدينة']
for i in range(len(customer_names)):
    email = f"{customer_names[i].lower()}@example.com"
    cursor.execute("INSERT INTO customers VALUES (?,?,?,?)", (i+1, customer_names[i], cities[i], email))

# إدخال بيانات الطلبات
order_dates = ['2023-01-15', '2023-02-10', '2023-03-05', '2023-04-20', '2023-05-12']
for i in range(len(order_dates)):
    customer_id = np.random.randint(1, len(customer_names)+1)
    total = round(np.random.uniform(50, 1000), 2)
    cursor.execute("INSERT INTO orders VALUES (?,?,?,?)", (i+1, customer_id, order_dates[i], total))

conn.commit()
print("تم إعداد قاعدة البيانات بنجاح.")
```
