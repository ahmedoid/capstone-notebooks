````python
# TODO: شغّل هذه الخلية لإنشاء بيانات التدريب
import pandas as pd
import numpy as np

np.random.seed(42)

# إنشاء جدول المنتجات
num_products = 20
product_ids = np.arange(1, num_products + 1)
product_names = [f'Product_{i}' for i in product_ids]
categories = np.random.choice(['Electronics', 'Clothing', 'Home'], size=num_products)
prices = np.random.uniform(10.0, 500.0, size=num_products).round(2)
products_df = pd.DataFrame({
    'product_id': product_ids,
    'name': product_names,
    'category': categories,
    'price': prices
})

# إنشاء جدول العملاء
num_customers = 10
customer_ids = np.arange(1, num_customers + 1)
customer_names = [f'Customer_{i}' for i in customer_ids]
cities = np.random.choice(['New York', 'London', 'Paris', 'Tokyo', 'Dubai'], size=num_customers)
customers_df = pd.DataFrame({
    'customer_id': customer_ids,
    'name': customer_names,
    'city': cities
})
````
