# مختبر: تصفية البيانات باستخدام WHERE

في هذا المختبر، ستتعلم استخدام جملة WHERE في SQL لتصفية البيانات من جداول قاعدة بيانات SQLite. ستقوم بكتابة استعلامات مختلفة لاستخراج صفوف معينة وفق شروط.

```python
import sqlite3
import pandas as pd
import numpy as np
from IPython.display import display

np.random.seed(42)

# إنشاء اتصال بقاعدة بيانات في الذاكرة
conn = sqlite3.connect(':memory:')

# إنشاء جدول المنتجات
conn.execute('''
    CREATE TABLE products (
        product_id INTEGER PRIMARY KEY,
        name TEXT,
        category TEXT,
        price REAL
    )
''')

# إنشاء جدول العملاء
conn.execute('''
    CREATE TABLE customers (
        customer_id INTEGER PRIMARY KEY,
        name TEXT,
        city TEXT
    )
''')

# إنشاء جدول الطلبات
conn.execute('''
    CREATE TABLE orders (
        order_id INTEGER PRIMARY KEY,
        customer_id INTEGER,
        product_id INTEGER,
        quantity INTEGER,
        order_date TEXT
    )
''')

# بيانات المنتجات
product_names = ['حاسوب', 'هاتف', 'لوحة مفاتيح', 'شاشة', 'قميص', 'بنطال', 'قلم', 'دفتر', 'لعبة فيديو', 'دمية']
categories = ['إلكترونيات', 'إلكترونيات', 'إلكترونيات', 'إلكترونيات', 'ملابس', 'ملابس', 'أدوات مكتبية', 'أدوات مكتبية', 'ألعاب', 'ألعاب']
prices = np.random.uniform(10, 300, 10).round(2)
for i in range(10):
    conn.execute("INSERT INTO products VALUES (?, ?, ?, ?)", (i+1, product_names[i], categories[i], prices[i]))
# تعديل سعر أحد منتجات الإلكترونيات ليقع في نطاق 50-200 للاختبار
conn.execute("UPDATE products SET price = 150 WHERE product_id = 1")

# بيانات العملاء
customer_names = ['أحمد', 'سارة', 'محمد', 'فاطمة', 'عمر', 'نورة', 'علي', 'مريم']
cities = ['الرياض', 'جدة', 'الرياض', 'جدة', 'الدمام', 'الرياض', 'مكة', 'جدة']
for i in range(8):
    conn.execute("INSERT INTO customers VALUES (?, ?, ?)", (i+1, customer_names[i], cities[i]))

# بيانات الطلبات
for i in range(15):
    customer_id = np.random.randint(1, 9)
    product_id = np.random.randint(1, 11)
    quantity = np.random.randint(1, 5)
    day = np.random.randint(1, 31)
    order_date = f'2025-01-{day:02d}'
    conn.execute("INSERT INTO orders VALUES (?, ?, ?, ?, ?)", (i+1, customer_id, product_id, quantity, order_date))
# إضافة طلب محدد لضمان وجود نتيجة في اختبار الوحدة
conn.execute("INSERT INTO orders VALUES (16, 5, 1, 3, '2025-01-10')")

conn.commit()
print("تم إنشاء قاعدة البيانات وإدخال البيانات بنجاح.")
# عرض عينة من كل جدول
print("\nجدول المنتجات:")
display(pd.read_sql_query("SELECT * FROM products LIMIT 3", conn))
print("\nجدول العملاء:")
display(pd.read_sql_query("SELECT * FROM customers LIMIT 3", conn))
print("\nجدول الطلبات:")
display(pd.read_sql_query("SELECT * FROM orders LIMIT 3", conn))
```
