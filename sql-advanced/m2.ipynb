# مختبر SQL المتقدم - تقنيات الاستعلامات الفرعية و UNION و CASE والفهارس
في هذا المختبر سنستخدم مكتبة sqlite3 في بايثون للتفاعل مع قاعدة بيانات SQLite مؤقتة. سنقوم بإنشاء جداول وإدخال بيانات عشوائية باستخدام numpy (ببذرة 42 لضمان التكرار)، ثم نكتب استعلامات SQL لتطبيق المفاهيم المتقدمة: الاستعلامات الفرعية، UNION، CASE، والفهارس، وأخيرًا مشروع تحليلي متكامل.
تأكد من إكمال جميع تمارين TODO.

```python
import sqlite3
import pandas as pd
import numpy as np
import random
np.random.seed(42)
random.seed(42)
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()
# إنشاء الجداول
cursor.execute('''
    CREATE TABLE customers (
        id INTEGER PRIMARY KEY,
        name TEXT NOT NULL,
        city TEXT NOT NULL,
        signup_date DATE NOT NULL
    )
''')
cursor.execute('''
    CREATE TABLE products (
        id INTEGER PRIMARY KEY,
        name TEXT NOT NULL,
        price REAL NOT NULL,
        category TEXT NOT NULL
    )
''')
cursor.execute('''
    CREATE TABLE orders (
        id INTEGER PRIMARY KEY,
        customer_id INTEGER REFERENCES customers(id),
        product_id INTEGER REFERENCES products(id),
        quantity INTEGER NOT NULL,
        order_date DATE NOT NULL
    )
''')
conn.commit()
```

```python
# توليد بيانات العملاء
customer_names = [f'Customer_{i}' for i in range(1, 31)]
cities = ['الرياض', 'جدة', 'الدمام', 'مكة', 'المدينة', 'الخبر']
customer_cities = [random.choice(cities) for _ in range(30)]
base_date = np.datetime64('2022-01-01')
signup_dates = base_date + np.random.randint(0, 365, size=30).astype('timedelta64[D]')
customers_data = list(zip(customer_names, customer_cities, signup_dates))
cursor.executemany('INSERT INTO customers (name, city, signup_date) VALUES (?, ?, ?)', customers_data)

# توليد بيانات المنتجات
product_names = [f'Product_{i}' for i in range(1, 16)]
prices = np.round(np.random.uniform(10, 200, size=15), 2)
categories = ['إلكترونيات', 'ملابس', 'طعام', 'أثاث', 'كتب']
product_categories = [random.choice(categories) for _ in range(15)]
products_data = list(zip(product_names, prices, product_categories))
cursor.executemany('INSERT INTO products (name, price, category) VALUES (?, ?, ?)', products_data)

# توليد بيانات الطلبات
num_orders = 150
order_customer_ids = np.random.randint(1, 31, size=num_orders)
order_product_ids = np.random.randint(1, 16, size=num_orders)
quantities = np.random.randint(1, 6, size=num_orders)
order_base = np.datetime64('2023-01-01')
order_dates = order_base + np.random.randint(0, 365, size=num_orders).astype('timedelta64[D]')
orders_data = list(zip(order_customer_ids, order_product_ids, quantities, order_dates))
cursor.executemany('INSERT INTO orders (customer_id, product_id, quantity, order_date) VALUES (?, ?, ?, ?)', orders_data)

conn.commit()
print('تم إدخال البيانات بنجاح.')
```
