```python
# مشروع ختامي: مولّد أوصاف تسويقية عربية
في هذا المشروع، ستبني مولّدًا للوصف التسويقي باللغة العربية لمنتجات متجر، باستخدام LangChain ونموذج DeepSeek.

**الهدف**: إنشاء أداة تولّد وصفًا قصيرًا وجذابًا لكل منتج، مع التحكم في نبرة الوصف (رسمية أو ودّية) لتتناسب مع الجمهور المستهدف.

ستتعلم في هذا المشروع:
- كيفية استدعاء نموذج لغوي كبير عبر LangChain.
- تصميم مطالبات (prompts) فعّالة.
- إضافة تحكم في الأسلوب عبر وسائط الدالة.
```

```python
## 1. إعداد البيئة
قم باستيراد المكتبات اللازمة. مفتاح API الخاص بـ DeepSeek محقون في المتغير البيئي `DEEPSEEK_API_KEY` تلقائيًا. لا تقم بإدخاله يدويًا.
```

```python
import os
from langchain_openai import ChatOpenAI

# التحقق من وجود المفتاح (اختياري)
if "DEEPSEEK_API_KEY" not in os.environ:
    raise ValueError("لم يتم العثور على DEEPSEEK_API_KEY في متغيرات البيئة")
```

```python
# تعريف النموذج
llm = ChatOpenAI(
    model="deepseek-chat",
    base_url="https://api.deepseek.com/v1",
    api_key=os.environ["DEEPSEEK_API_KEY"]
)
# اختبار سريع
print(llm.invoke("قل مرحباً").content)
```

```python
## 2. بيانات المنتجات
لدينا قائمة من المنتجات، كل منتج معرّف بقاموس يحوي اسم المنتج، الفئة، والسعر. سنستخدم هذه البيانات لتوليد الأوصاف.

في خلية لاحقة، ستقوم بإنشاء الدالة التي تتولى كل منتج وتولّد وصفًا تسويقيًا.
```

```python
def describe_product(product):
    # قم ببناء المطالبة هنا
    prompt = f""" ... """
    
    # استدعاء النموذج
    response = llm.invoke(prompt)
    
    # إرجاع النص الناتج
    return response.content
```

```python
## 4. مهمة 2 – إضافة وسيط النبرة
الآن نريد التحكم في أسلوب الوصف عبر وسيط `نبرة` (tone). يمكن أن تكون `"رسمية"` أو `"ودّية"` (أضف أي أنماط أخرى إن أردت).

**المطلوب**:
- عدّل دالة `describe_product` بحيث تستقبل وسيطًا إضافيًا `tone` بقيمة افتراضية `"رسمية"`.
- غيّر المطالبة لتشمل توجيهًا حول النبرة المطلوبة. مثلاً: إذا كانت النبرة "ودّية"، استخدم لغة دافئة وغير رسمية، وإذا كانت "رسمية" استخدم لغة مهنية.

أكمل الخلية أدناه.
```
