```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
## توليد البيانات التخليقيةسنقوم بتوليد 120 سجلًا لطلاب افتراضيين. كل سجل يحتوي على:- `study_hours`: عدد ساعات الدراسة (بين 0 و 10)- `attendance`: نسبة الحضور (بين 0 و 1)- `passed`: هل نجح الطالب؟ (1 = نعم، 0 = لا)سنستخدم قاعدة بسيطة مع إضافة بعض الضوضاء لجعل المسألة غير حتمية تمامًا.
```

```python
# إعداد البيانات
np.random.seed(42)
n = 120
study_hours = np.random.uniform(0, 10, n)
attendance = np.random.uniform(0, 1, n)

# قاعدة بسيطة: ينجح إذا درس أكثر من 5 ساعات وكان الحضور أعلى من 70%
passed = ((study_hours > 5) & (attendance > 0.7)).astype(int)

# إضافة ضوضاء عشوائية (قلب 5% من النتائج)
noise_idx = np.random.choice(n, size=int(0.05 * n), replace=False)
passed[noise_idx] = 1 - passed[noise_idx]

# إنشاء DataFrame
df = pd.DataFrame({'study_hours': study_hours, 'attendance': attendance, 'passed': passed})
print('تم إنشاء البيانات بنجاح.')
```
