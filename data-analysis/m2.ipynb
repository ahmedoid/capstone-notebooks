## 4. اختبار كاي تربيع (Chi-Square Test)

يُستخدم اختبار كاي تربيع لفحص العلاقة بين متغيرين فئويين. سنقوم بتحليل بيانات مسح عن الجنس (ذكر/أنثى) وحالة التدخين (مدخن/غير مدخن) لاختبار استقلالية المتغيرين.

```python
# إنشاء جدول التوافق (Contingency Table)
# سنقوم بتوليد بيانات عشوائية
np.random.seed(123)  # بذرة منفصلة للتوضيح
n_total = 300
# توليد عمود الجنس
gender = np.random.choice(['Male', 'Female'], size=n_total, p=[0.5, 0.5])
# توليد عمود التدخين مع ارتباط جزئي (لجعل المثال مثيرًا)
smoking = []
for g in gender:
    if g == 'Male':
        smoking.append(np.random.choice(['Yes', 'No'], p=[0.3, 0.7]))
    else:
        smoking.append(np.random.choice(['Yes', 'No'], p=[0.2, 0.8]))
smoking = np.array(smoking)

df = pd.DataFrame({'Gender': gender, 'Smoking': smoking})
print('عينة من البيانات:')
print(df.head(10))

# إنشاء جدول التوافق
observed = pd.crosstab(df['Gender'], df['Smoking'])
print('\nجدول التوافق (Observed):')
print(observed)

# TODO: قم بإجراء اختبار كاي تربيع باستخدام stats.chi2_contingency على جدول observed
# استخرج قيمة chi2, p, dof, expected
chi2, p, dof, expected = None, None, None, None

print('\nChi-square statistic:', chi2)
print('p-value:', p)
print('Degrees of freedom:', dof)
print('Expected frequencies table:')
print(expected)

# TODO: فسر النتيجة عند alpha=0.05
```

## خلاصة وتفسير p-value

في جميع الاختبارات السابقة، استخدمنا القيمة الاحتمالية (p-value) لاتخاذ قرار بشأن الفرضية الصفرية. قاعدة عامة:
- إذا كانت p-value أقل من مستوى الدلالة (مثلاً 0.05)، نرفض الفرضية الصفرية ونعتبر النتيجة ذات دلالة إحصائية (statistically significant).
- إذا كانت p-value أكبر من أو تساوي مستوى الدلالة، لا نرفض الفرضية الصفرية (أي لا يوجد دليل كاف لرفضها).

تذكر أن عدم رفض الفرضية الصفرية لا يعني أنها صحيحة بالضرورة، بل يعني أن البيانات لا تقدم دليلًا كافيًا ضدها.
