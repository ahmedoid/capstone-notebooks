## الخطوة 5: تطبيق على بيانات تركيبية مع نموذج k-NN

سنستخدم مكتبة Scikit-learn لتوليد بيانات تصنيف ثنائي، وتقسيمها إلى تدريب واختبار، ثم تدريب مصنف k-NN. بعد ذلك سنقيم النموذج باستخدام الدوال التي بنيناها.

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

np.random.seed(42)

# توليد بيانات تصنيف ثنائي
X, y = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# تدريب مصنف k-NN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

print('تم تدريب النموذج والحصول على التنبؤات.')
```
