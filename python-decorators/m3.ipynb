## 1. تزيين دوال ذات وسائط باستخدام *args و **kwargs

```python
import functools
import time
import numpy as np

np.random.seed(42)

def log_args(func):
    # TODO: أكمل الدالة wrapper بحيث تطبع اسم الدالة والوسائط (args) والوسائط المسماة (kwargs) قبل تنفيذ func
    def wrapper(*args, **kwargs):
        # TODO: اكتب كود الطباعة هنا
        result = func(*args, **kwargs)
        return result
    return wrapper
```

```python
@log_args
def add(a, b):
    return a + b

print(add(3, 5))
```

## 2. إرجاع القيم من الديكوريتر

```python
def timing_decorator(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        # TODO: قم بتنفيذ func(*args, **kwargs) واحفظ النتيجة في المتغير result
        result = None
        end = time.time()
        print(f'{func.__name__} took {end - start:.4f} seconds')
        # TODO: أعد النتيجة المحفوظة
        return result
    return wrapper
```

```python
@timing_decorator
def slow_add(a, b):
    time.sleep(0.1)
    return a + b

print(slow_add(10, 20))
```

## 3. الحفاظ على البيانات الوصفية مع functools.wraps

```python
# ديكوريتر بدون wraps
def no_wraps_decorator(func):
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs)
    return wrapper

@no_wraps_decorator
def my_func():
    'هذه دالة my_func الأصلية'
    return 42

print('بدون wraps:')
print(my_func.__name__)
print(my_func.__doc__)

# TODO: عرف دالة with_wraps_decorator مع إضافة @functools.wraps(func) أعلى wrapper
def with_wraps_decorator(func):
    # TODO: أضف @functools.wraps(func) هنا
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs)
    return wrapper

@with_wraps_decorator
def another_func():
    'دالة أخرى'
    return 7

print('باستخدام wraps:')
print(another_func.__name__)
print(another_func.__doc__)
```

## 4. مصانع الديكوريترز: تمرير وسائط إلى الديكوريتر

```python
def repeat(n):
    # مصنع ديكوريتر يأخذ عدد التكرارات n
    def decorator(func):
        # TODO: عرف wrapper الذي يكرر تنفيذ func عدد n مرات ويعيد آخر نتيجة
        def wrapper(*args, **kwargs):
            # TODO: قم بإنشاء حلقة for لتكرار func(*args, **kwargs) عدد n مرة وحفظ النتيجة في result
            result = None
            return result
        return wrapper
    return decorator
```

```python
@repeat(3)
def greet(name):
    print(f'Hello {name}')
    return f'Hi {name}'

result = greet('Alice')
print(result)
```

## تمرين: بناء ديكوريتر Retry

في هذا التمرين، ستقوم ببناء ديكوريتر retry يعيد محاولة تنفيذ دالة عند حدوث استثناء، مع حد أقصى لعدد المحاولات وتأخير زمني. استخدم np.random.rand() لمحاكاة الفشل العشوائي.

```python
def retry(max_retries=3, delay=0.5):
    'ديكوريتر يعيد محاولة تنفيذ الدالة إذا حدث استثناء'
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            for attempt in range(max_retries):
                try:
                    # TODO: قم بتنفيذ func(*args, **kwargs) وأعد النتيجة مباشرة
                    pass
                except Exception as e:
                    print(f'Attempt {attempt+1} failed with {e}')
                    if attempt == max_retries - 1:
                        raise
                    time.sleep(delay)
        return wrapper
    return decorator
```

```python
@retry(max_retries=3, delay=0.1)
def unstable_function(x):
    # تحاكي فشل بنسبة 70% باستخدام بذرة عشوائية
    if np.random.rand() < 0.7:
        raise ValueError(f'Random failure for x={x}')
    return x * 2

np.random.seed(42)
for i in range(5):
    try:
        result = unstable_function(i)
        print(f'Success: {result}')
    except ValueError as e:
        print(f'Final failure: {e}')
```
