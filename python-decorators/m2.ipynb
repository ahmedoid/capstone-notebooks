## أول ديكوريتر: البناء والصيغة (First Decorator)

الديكوريتر هو دالة تستقبل دالة وتُعيد دالة جديدة تحوي سلوكًا إضافيًا. تستخدم الصيغة @ لتطبيق الديكوريتر بسهولة.

```python
# بناء ديكوريتر بسيط
def simple_decorator(func):
    def wrapper(*args, **kwargs):
        # TODO: اطبع رسالة قبل استدعاء func
        pass
        result = func(*args, **kwargs)
        # TODO: اطبع رسالة بعد استدعاء func
        pass
        return result
    return wrapper

@simple_decorator
def say_hello(name):
    print(f"Hello, {name}")

say_hello("Ali")
```

```python
# مثال آخر: ديكوريتر على دالة تُرجع قيمة
@simple_decorator
def square(x):
    return x * x

print(square(4))
```

## تمرين: بناء Logger Decorator

في هذا التمرين ستقوم ببناء ديكوريتر يقوم بتسجيل (Logging) استدعاء الدالة: يطبع اسم الدالة، الوسائط (arguments)، والنتيجة (return value).

```python
# تمرين: بناء logger decorator
def logger(func):
    def wrapper(*args, **kwargs):
        # TODO: اطبع اسم الدالة والوسائط المستخدمة
        pass
        result = func(*args, **kwargs)
        # TODO: اطبع النتيجة التي أُعيدت
        pass
        return result
    return wrapper

@logger
def add(a, b):
    return a + b

@logger
def greet(name):
    return f"Hello, {name}"

print(add(3, 5))
print(greet("Sarah"))
```

```python
# اختبار إضافي مع وسائط مختلفة
@logger
def power(base, exp=2):
    return base ** exp

print(power(3))
print(power(2, exp=3))
```
